```python
import pandas as pd
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns

import os
import sys

from pathlib import Path

%matplotlib inline

from IPython.display import display, HTML
from weasyprint import HTML as pdf_c
```

```python
# Name of file to import raw data from (from bloomberg)
filename = "ALK Inhibitors Market vRH.xlsx"

# Dynamically create filepath independent of system
path = Path(os.getcwd())
filepath = os.path.abspath(os.path.join(path.parent, "imports", filename))

# appending to sys.path to allow for relative import of general functions
sys.path.append(os.path.abspath(path.parent.parent))
from drug_curves import *
```

```python
# Importing in data
raw_data = pd.read_excel(filepath, sheet_name="BB New", header=[0, 1, 2, 3])
raw_data.columns = multi_index_column_rename(raw_data.columns)

# Global Important Variables
DRUGS = ["Xalkori", "Zykadia", "Alecensa"]
DRUGS_M = ["Xalkori (crizotinib)", "Zykadia (ceritinib)", "Alecensa (alectinib)"]
UNITS = {
    "volume": "Integrated Units",
    "mbs": "Integrated Mbs"
}
```

# Organizing Data

```python
### Examinine that all date columns match
# Note -- df.duplicated was not working for some reason
idx = pd.IndexSlice[:, :, :, "Date"]
assert raw_data.loc[:, idx].apply(lambda row: row.duplicated(keep=False).all(), axis=1).sum() == raw_data.shape[0]

# Assert "Date" is a timestmap/datetime dtype
assert pd.core.dtypes.common.is_datetime_or_timedelta_dtype(raw_data.loc[:, idx].iloc[:, 0])

# Push the first "Date Column"
raw_data.set_index(raw_data.loc[:, idx].iloc[:, 0], inplace=True)
raw_data.index.name = "Date"


# Dropping "Date" Columns and ["Blank"] columns [if any]
raw_data.drop("Date", axis=1, level=3, inplace=True)
for level in raw_data.columns:
    if "[Blank]" in level:
        raw_data.drop("[Blank]", axis=1, level=3, inplace=True)
        break

# Dropping rows earlier than 1/31/2013
raw_data = raw_data.loc[raw_data.index >= pd.Timestamp(year=2013, month=1, day=31), :].copy()

# Dropping empty columns
columns_to_drop = []
for column in raw_data:
    if raw_data.loc[:, column].isnull().all():
        columns_to_drop.append(column)
raw_data.drop(columns_to_drop, axis=1, inplace=True)

# Changing column names to match pricerx db
mapper = {
    "Pfizer Us Pharm": "Pfizer",
    "Genentech Inc": "Roche",
         }
uncap_molecule = {
    "Xalkori (Crizotinib)": "Xalkori (crizotinib)",
    "Zykadia (Ceritinib)": "Zykadia (ceritinib)",
    "Alecensa (Alectinib)": "Alecensa (alectinib)",
}
raw_data.rename(mapper, axis=1, level=1, inplace=True)
raw_data.rename(mapper, axis=1, level=3, inplace=True)
raw_data.rename(uncap_molecule, axis=1, level=0, inplace=True)

# Visualizing our new modified df
raw_data.head()
```

# Data Validation

```python
# These assertions check that we are capturing the entire drug volume (we breakout by manuf.)
for drug in DRUGS_M:
    all_manufacturers = raw_data.loc[:, (drug, "All", slice(None), slice(None))].columns.get_level_values(3).unique()
    for manuf in all_manufacturers:
        total = raw_data.loc[:, (drug, "All", UNITS["volume"], manuf)].fillna(0)
        s = raw_data.loc[:, (drug, manuf, UNITS["volume"], slice(None))].sum(axis=1)
        if not (total == s).all():
            raise ValueError(f"Units for {drug}, {manuf} does not match total")

# Replacing negative numbers with 0s
raw_data.mask(raw_data < 0, 0, inplace=True)
```

```python
# Creating 1) validation reports 2) excel breakout tables 3) "tidy" df for later tables
pricerx_df = pricerx_df_gen(DRUGS)
pricerx_df["WAC/Package"] = (pricerx_df["Price"] / pricerx_df["Package"]).astype(float)
frames = {}
calculated_df = pd.DataFrame()
report = pd.DataFrame()

for i, drug in enumerate(DRUGS_M):
    frames[DRUGS[i]] = {}
    all_manufacturers = raw_data.loc[:, (drug, "All", slice(None), slice(None))].columns.get_level_values(3).unique()
    for manuf in all_manufacturers:
        # Getting data for Units(Total), Retail, and MBS
        base = raw_data.loc[:, (drug, "All", slice(None), manuf)].copy()
        base.columns = base.columns.get_level_values(2)
        
        # Getting data for Units -- broken out by strength
        str_breakdown = raw_data.loc[:, (drug, manuf, slice(None), slice(None))].copy()
        
        # Adjusting strength column names to match pricerx database
        col = list(str_breakdown.columns.get_level_values(3))
        new_cols = []
        new_cols_wac = []
        for c in col:
            cutoff = c.lower().find("m")
            replacement = f"{c[:cutoff]} {c[cutoff:]}".upper()
            new_cols.append(replacement)
        str_breakdown.columns = new_cols
        
        # Creating the calculated_df (and math df)
        manufacturer_df = pd.concat([str_breakdown, base], axis=1)
        temp = pd.DataFrame()
        
        # Adding the WAC/ Package Columns
        for c in new_cols:
            name = f"{c} WAC/Package"
            new_cols_wac.append(name)
            prices = pricerx_df.loc[
                (pricerx_df["Drug"] == DRUGS[i]) &
                (pricerx_df["Manufacturer"] == manuf) &
                (pricerx_df["Strength"] == c), "WAC/Package"]
            manufacturer_df[name] = prices
            temp[c] = manufacturer_df[c] * prices
        
        # Adding the rest of the calculated columns
        manufacturer_df["Pills/Rx"] = manufacturer_df["Retail Quantity"] / manufacturer_df["Retail Count"]
        manufacturer_df["TTM Pills/Rx"] = manufacturer_df["Pills/Rx"][0:12].mean()
        manufacturer_df["WA WAC/Package"] = temp.sum(axis=1)/manufacturer_df[UNITS["volume"]]
        manufacturer_df["Total Rx"] = manufacturer_df[UNITS["volume"]] / manufacturer_df["Pills/Rx"]
        manufacturer_df["Total WAC Sales"] = (manufacturer_df["WA WAC/Package"] *
                                              manufacturer_df["Pills/Rx"] *
                                              manufacturer_df["Total Rx"]).round(2)
        manufacturer_df["TRx"] = manufacturer_df[UNITS["volume"]] / manufacturer_df["TTM Pills/Rx"]
        manufacturer_df["WAC/TRx"] = manufacturer_df["WA WAC/Package"] * manufacturer_df["TTM Pills/Rx"]
        manufacturer_df["Difference"] = (manufacturer_df["Total WAC Sales"] - manufacturer_df[UNITS["mbs"]]).round(2)
        manufacturer_df["Percentage Difference"] = (manufacturer_df["Difference"] / manufacturer_df[UNITS["mbs"]]).round(2)
        column_order = (new_cols + [UNITS["volume"], "Retail Quantity", "Retail Count", "Pills/Rx", "TTM Pills/Rx"] +
                        new_cols_wac + ["WA WAC/Package", "Total Rx", "TRx", "WAC/TRx",
                                        "Total WAC Sales", UNITS["mbs"], 
                                        "Difference", "Percentage Difference"])
        # Re-ordering the columns
        manufacturer_df = manufacturer_df[column_order].copy()
        
        # Adding frame to frames (for excel)
        frames[DRUGS[i]][manuf] = manufacturer_df
        
        # Adding to report
        report[("{}-{}".format(drug, manuf))] = manufacturer_df["Percentage Difference"]
        
        #Tidying (kinda) the data and adding to calculated df
        calculated_df_temp = pd.DataFrame()
        
        calculated_df_temp["Volume"] = manufacturer_df["TRx"]
        calculated_df_temp["PriceRx"] = manufacturer_df["WAC/TRx"]
        calculated_df_temp["WAC"] = manufacturer_df["Total WAC Sales"]
        calculated_df_temp["Volume_type"] = "TRx"
        calculated_df_temp["Price_type"] = "WAC/TRx"
        calculated_df_temp["Drug"] = drug
        calculated_df_temp["Manufacturer"] = manuf
                
        calculated_df = calculated_df.append(calculated_df_temp)

def color_negative_red(x):
    y = float(x[:x.find("%")])
    if y > 0 or y < 0:
        color = "red"
    else:
        color = "black"
    return 'color: {}'.format(color)

# Printing out the report
report = report.applymap(lambda val: "{:.0%}".format(val))
report.index = report.index.date
report = report.head(12).style.applymap(color_negative_red)
report
```
